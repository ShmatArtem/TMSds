```python
df = pd.read_csv("train.csv")
df.sample(10)
```

```python
df.info()
```

```python
# Половозрастная структура
a = df['Age'][df.Sex == 'male'].agg(['mean', 'min', 'max'])
b = df['Age'][df.Sex == 'female'].agg(['mean', 'min', 'max'])

print(f'Средний возраст мужчины: {a[0]}')
print(f'Средний возраст женщины: {b[0]}')

print(f'Минимальный возраст мужчины: {a[1]}')
print(f'Максимальный возраст женщины: {b[1]}')

print(f'Минимальный возраст мужчины: {a[2]}')
print(f'Минимальный возраст женщины: {b[2]}')
```

```python
# Создал колонку возратсных групп
df['AgeGroup'] = range(len(df))

for i in range(len(df)):
  if df.Age[i] < 18:
    df['AgeGroup'][i] = 'Child'
  elif df.Age[i] < 60 and df.Age[i] >= 18:
    df['AgeGroup'][i] = 'Adult'
  elif df.Age[i] > 60:
    df['AgeGroup'][i] = 'Old man'
  else:
    df['AgeGroup'][i] = None


df.sample(10)
```

```python
# Вероятность выживаемости мужчины по его возрастной группе
a = df['Age'][df.Sex == 'male'].groupby(df.AgeGroup).agg('count')
b= df['Age'][df.Sex == 'male'][df.Survived == 1].groupby(df.AgeGroup).agg('count')
b / a * 100
```

```python
# Вероятность выживаемости женщины по её возрастной группе
c = df['Age'][df.Sex == 'female'].groupby(df.AgeGroup).agg('count')
d= df['Age'][df.Sex == 'female'][df.Survived == 1].groupby(df.AgeGroup).agg('count')
d / c * 100
```

```python
#Половозрастная структура по выжившим и нет
print(df[['Survived', 'Age', 'Sex']][df.Survived == 1].groupby('Sex').agg({'Age': ['mean', 'median']}))
df[['Survived', 'Age', 'Sex']][df.Survived == 0].groupby('Sex').agg({'Age': ['mean','median']})
```

```python
# Количество пустых значений
df.isna().sum()
```

```python
# Средняя стоимость билета
mean_fare = df['Fare'].mean()
print(f'Средняя цена билета: {mean_fare}')
```

```python
# Средняя стоимость билета выжившего
mean_fare = df['Fare'][df.Survived == 1].mean()
print(f'Средняя цена билета: {mean_fare}')
```

```python
# Влияние класса на выживаемость
a = df['Survived'][df.Survived == 1].groupby(df.Pclass).count()
b = df['Survived'].groupby(df.Pclass).count()
a / b * 100
```

```python
# Сопоставить колонки Fare и Embarked
df[['Fare', 'Embarked']].groupby('Embarked').agg({'Fare' : ['mean', 'max', 'mean']})
```

```python
# Количество выживжих в зависимости от порта
df[['Survived', 'Embarked']].groupby('Embarked').agg('sum')
```

```python
# Зависимость выживаемости от порта
a = df[['Survived', 'Embarked']][df.Survived == 1].groupby('Embarked').agg('count')
b= df[['Survived', 'Embarked']].groupby('Embarked').agg('count')
a / b * 100
```

```python
# Количество выживших младенцев (до 1 года)
alive = df['Name'][df.Age < 1][df.Survived == 1].count()
dead = df['Name'][df.Age < 1][df.Survived == 0].count()

print(f'{alive} младенцев были спасены, погибших {dead}')
```

```python
sorted_df = df.sort_values('Name').head(15)
sorted_df
```

```python
# Выделяем стобец с фамилиями
df['LastName'] = df['Name'].apply(lambda x: x.split(',')[0])
# Отбираем группы с количеством одинаковых фамилий > 1
relatives = df.groupby(['LastName', 'Ticket']).filter(lambda x: len(x) > 1)
families = relatives[(df.SibSp > 0) | (df.Parch > 0)]
# Находим количество семей
num_families = families.groupby(['LastName', 'Ticket']).ngroups
num_families
```
