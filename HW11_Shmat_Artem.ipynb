```python
import numpy as np
import matplotlib.pyplot as plt


def matrix_func(a, b, c: np.ndarray) -> str:
  """
  Returns the results of addition A+B, substraction B-A, multiplication A*C and A*B*C

  Args:
    a (np.ndarray): matrix A to perform operations above.
    b (np.ndarray): matrix B to perform operations above.
    c (np.ndarray): matrix C to perform operations above.

  Returns:
    result (str): the f-string containing the results of the matrix operations
  """
  addition = a + b
  substruction = b - a
  multiplication_ab = np.dot(a,b)
  multiplication_ac = np.dot(a,c)
  multiplication_abc = np.dot(multiplication_ab, c)

  return f'''A + B =
{addition}

B - A =
{substruction}

A * C =
{multiplication_ac}

A * B * C =
{multiplication_abc}'''
```

```python
# Task 1
a = np.array([
    [1,2,4],
    [7,5,-2],
    [-2,1,3]
])

b = np.array([
    [-1,0,3],
    [2,-6,1],
    [-7,1,5]
])

c = np.array([
    [-2,3],
    [4,0],
    [7,-1]
])

print(matrix_func(a,b,c))
```

```python
# Matrix filter
def convolution(image: np.ndarray, kernel: np.ndarray) -> np.ndarray:
  image_height, image_width, image_channels = image.shape
  kernel_height, kernel_width = kernel.shape

  convolution_matrix = np.zeros((image_height - kernel_height + 1, image_width - kernel_width + 1, image_channels))

  for channel in range(image_channels):
    for height in range(image_height - kernel_height + 1):
      for width in range(image_width - kernel_width + 1):
        convolution_matrix[height, width, channel] = np.sum(image[height:height + kernel_height, width:width + kernel_width, channel] * kernel).clip(0,255)

  return convolution_matrix.astype(np.uint8)
```

```python
kernel = np.array([
    [-1,-1,-1],
    [-1,9,-1],
    [-1,-1,-1]
])

image = plt.imread('electric-guitar.webp')

filtered_image = convolution(image, kernel)

plt.figure(figsize=(15, 10))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Origin image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(filtered_image)
plt.title('Filtered image')
plt.axis('off')

plt.show()
```
